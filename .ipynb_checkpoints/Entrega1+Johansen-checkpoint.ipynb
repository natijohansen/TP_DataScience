```python
# Importamos librerias a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
# Lectura del csv de Aerolineas y Vuelos
df_air = pd.read_csv('Airlines.csv', sep=',')
df_air.head()
```

1. **Airline:** Diferentes tipos de aerolineas comerciales
2. **Flight:** Número de Vuelo
3. **Airport From:** Aeropuesto de salida
4. **Airport To:** Aeropuerto de llegada
5. **DayOfWeek:** Día de la semana en el que se realiza el vuelo
6. **Time:** Duración del vuelo (minutos)
7. **Length:** Distancia
8. **Delay:** booleano si el vuelo tuvo demoras o no

**Data Features - Airports.csv**

1. **Airport:** Nombre completo del aeropuerto
2. **Cod_Airport:** Código de identificacion del aeropuerto
3. **Desc_Airport:** Nombre corto del Aeropuerto
4. **Loc:** Localización del Aeropuerto

#### 2.1 Data Wrangling

En la distribución de las variables númericas hemos identificado que existen vuelos con distancia 0 (Length = 0). Procedemos a eliminar estos registros 

```python
df_air.drop(df_air[df_air.Length == 0].index, inplace=True)
df_air.shape
```

Creamos una nueva columna con agrupacion de los tiempos de vuelos

```python
df_air.loc[df_air['Time'] <= 500, 'TimeGroups'] = '<= 500'
df_air.loc[df_air['Time'].between(500, 801, inclusive='neither'), 'TimeGroups'] = '501 - 800'
df_air.loc[df_air['Time'].between(800, 1101, inclusive='neither'), 'TimeGroups'] = '801 - 1100'
df_air.loc[df_air['Time'] > 1100, 'TimeGroups'] = '> 1100'

df_air
```

Creamos una nueva columna con la agrupación de las distancias

```python
df_air.loc[df_air['Length'] <= 80, 'LengthGroups'] = '<= 80'
df_air.loc[df_air['Length'].between(80, 141, inclusive='neither'),  'LengthGroups'] = '81 - 140'
df_air.loc[df_air['Length'].between(140, 201, inclusive='neither'),  'LengthGroups'] = '141 - 200'
df_air.loc[df_air['Length'] > 200, 'LengthGroups'] = '> 200'
df_air
```

Incorporamos al dataset información sobre los aeropuestos

Merge Aeropuestos Origen

```python
# Merge Aeropuestos Origen
df_air = pd.merge(df_air, df_airport, how='left', left_on='AirportFrom', right_on='Cod_Airport')
df_air.head()
```

```python
# Renombramos columnas
df_air.rename(columns={"Airport":"Airport_From", "Cod_Airport":"Cod_AirportFrom", "Desc_Airport":"Desc_AirportFrom", "Loc":"Loc_From"}, inplace=True)
df_air.head()
```

Merge Aeropuestos Destinos

```python
# Merge Aeropuestos Destinos
df_air = pd.merge(df_air, df_airport, how='left', left_on='AirportTo', right_on='Cod_Airport')
df_air.head()
```

```python
# Renombramos columnas
df_air.rename(columns={"Airport":"Airport_To", "Cod_Airport":"Cod_AirportTo", "Desc_Airport":"Desc_AirportTo", "Loc":"Loc_To"}, inplace=True)
df_air.head()
```

Reemplazamos los valores "NaN" por "Other", valor que va a representar aquellos aeropuestos que no se encontraban en el dataset mergeado.  

```python
df_air['Cod_AirportFrom'].fillna("Other", inplace=True)
df_air['Cod_AirportTo'].fillna("Other", inplace=True)
df_air['Loc_From'].fillna("Other", inplace=True)
df_air['Loc_To'].fillna("Other", inplace=True)

df_air.head()
```

A continuación haremos un **Análisis de los retrasos considerando los Aeropuertos de Origen y Destino**, y luego la localización de dichos aeropuertos

Filtro dataframe con solo los vuelos que tuvieron delay

```python
delays = df_air['Delay'] == 1
df_delay = df_air[delays]
df_delay.head()
```

Agrupamos "Delay" segun AirportFrom and AirportTo

```python
# Agrupo "Delay" segun AirportFrom and AirportTo
f_airport = df_delay.groupby(by=['Cod_AirportFrom','Cod_AirportTo'])['id'].count()
f_airport
```

Creamos nuevo dataframe con el resultado de la agrupación

```python
df_aux1 = f_airport.unstack(level='Cod_AirportTo')
df_aux1.head()
```

```python
plt.figure(figsize = (42,22))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df_aux1, annot=True, cmap = cmap, fmt=".1f")
plt.show()
```

Repetimos el proceso para la Localización de los Aeropuestos

```python
# Agrupo "Delay" segun Loc_From and Loc_To
f_loc = df_delay.groupby(by=['Loc_From','Loc_To'])['id'].count()

df_aux3 = f_loc.unstack(level='Loc_To')
df_aux3.head()
```

```python
plt.figure(figsize = (30,15))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df_aux3, annot=True, cmap = cmap, fmt=".1f")
plt.show()
```
